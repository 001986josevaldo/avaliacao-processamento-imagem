```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

## Lista de Exercicios


```python
# Carregar imagem e converter BGR para RGB
img = cv2.imread('/content/imagens/moto.jpeg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img_rgb)
plt.axis('off')
```

```python
#  2 Imprima na tela os atributos: shape, size e dtype da imagem.
print(img.shape, img.size, img.dtype)
```

```python
# 3 Salve umacópia da sua imagem original em um formato de compressão diferente
#(ex: PNG para JPG)
cv2.imwrite('imagens/copia.png', img)
```

```python
# 4  Acesse as coordenadas centrais da imagem e exiba os valores R, G e B desse
#pixel específico.
h, w = img.shape[:2]
centro = img[h//2, w//2]
print("R,G,B:", centro[2], centro[1], centro[0])
```

```python
# 5
' Modifique uma região de 100 × 100 no canto superior para uma cor sólida eexiba o resultado.'
img_mod = img.copy()
img_mod[0:100, 0:100] = [0, 0, 255]
plt.imshow(cv2.cvtColor(img_mod, cv2.COLOR_BGR2RGB)); plt.axis('off')
```

```python
# 6  Realize um recorte (ROI) focando em um detalhe da sua foto e exiba apenas
# o recorte
roi = img[100:300, 100:300]
plt.imshow(cv2.cvtColor(roi, cv2.COLOR_BGR2RGB)); plt.axis('off')
```

```python
# 7  Separe os canais B, G e R da sua imagem utilizando funções do OpenCV
b, g, r = cv2.split(img)
```

```python
# 8 Exiba o canal Verde (G) isoladamente como uma imagem em escala de cinzentos
plt.imshow(g, cmap='gray'); plt.axis('off')
```

```python
# 9 Realize o merge dos canais, mas troque a posição do R com o B para criar um efeito visual.
merge = cv2.merge([r, g, b])
plt.imshow(cv2.cvtColor(merge, cv2.COLOR_BGR2RGB)); plt.axis('off')
```

```python
# 10  Converta a sua imagem original para tons de cinzento via software
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray, cmap='gray'); plt.axis('off')
```

```python
# 11  Converta a imagem para o espaço HSV e extraia apenas o canal de Saturação(S)
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
s = hsv[:,:,1]
plt.imshow(s, cmap='gray'); plt.axis('off')
```

```python
# 12  Aumente o brilho da imagem somando 50 unidades aos pixéis (trate o transbordo com cv2.add)
brilho = cv2.add(img, np.ones(img.shape, dtype=np.uint8)*50)
plt.imshow(cv2.cvtColor(brilho, cv2.COLOR_BGR2RGB)); plt.axis('off')
```

```python
# 13 Redimensione duas fotos para o mesmo tamanho e faça a mesclagem com addWeighted (pesos 0.5/0.5)
h, w = img.shape[:2]
img2 = cv2.resize(img, (w, h))
blend = cv2.addWeighted(img, 0.5, img2, 0.5, 0)
plt.imshow(cv2.cvtColor(blend, cv2.COLOR_BGR2RGB)); plt.axis('off')
```

```python
# 14  Implemente o negativo fotográfico da imagem em tons de cinzento.
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
neg = cv2.bitwise_not(gray)

plt.imshow(neg, cmap='gray')
plt.axis('off')
```

```python
# 15  Crie uma máscara binária circular e use-a para isolar uma parte da sua foto principal
h, w = img.shape[:2]
mask = np.zeros((h, w), dtype=np.uint8)

raio = min(h, w) // 4
cv2.circle(mask, (w//2, h//2), raio, 255, -1)

isolado = cv2.bitwise_and(img, img, mask=mask)

plt.imshow(cv2.cvtColor(isolado, cv2.COLOR_BGR2RGB))
plt.axis('off')
```

```python
# 16 Reduza a imagem original para 30% do tamanho inicial e exiba o novo shape.
res = cv2.resize(img, None, fx=0.3, fy=0.3)
print(res.shape)
```

```python
# 17 Calcule e imprima os valores de intensidade do histograma da sua imagem.

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
hist = cv2.calcHist([gray], [0], None, [256], [0,256])

print(hist.flatten().astype(int))
```

```python
# 18  Plote o histograma (gráfico) dos canais de cores separadamente no mesmo gráfico
plt.figure()

colors = ('b','g','r')
for i, c in enumerate(colors):
    hst = cv2.calcHist([img],[i],None,[256],[0,256])
    plt.plot(hst, color=c)

plt.xlim([0,256])
plt.show()
```

```python
# 19 Aplique a Equalização de Histograma para melhorar o contraste de uma imagem escura sua.

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
eq = cv2.equalizeHist(gray)

plt.imshow(eq, cmap='gray')
plt.axis('off')
```

```python
# 20 Aplique um limiar (Threshold) binário simples para separar o objeto do fundona sua foto.

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
_, th = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

plt.imshow(th, cmap='gray')
plt.axis('off')
```
