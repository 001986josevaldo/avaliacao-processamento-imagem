# Exercício 1 - Carregar imagem e converter BGR para RGB

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('/content/imagens/moto.jpeg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img_rgb)
plt.axis('off')
```

```python
# 2
print(img.shape, img.size, img.dtype)
```

```python
# 3
cv2.imwrite('imagens/copia.png', img)
```

```python
# 4
h, w = img.shape[:2]
centro = img[h//2, w//2]
print("R,G,B:", centro[2], centro[1], centro[0])
```

```python
# 5
img_mod = img.copy()
img_mod[0:100, 0:100] = [0, 0, 255]
plt.imshow(cv2.cvtColor(img_mod, cv2.COLOR_BGR2RGB)); plt.axis('off')
```

```python
# 6
roi = img[100:300, 100:300]
plt.imshow(cv2.cvtColor(roi, cv2.COLOR_BGR2RGB)); plt.axis('off')
```
