# Exercício 1 - Carregar imagem e converter BGR para RGB

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('/content/avaliacao-processamento-imagem/evidencia_janela.png')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img_rgb)
plt.axis('off')
```
